#局部線性嵌入法（Locally Linear Embedding ，LLE）


*   流行學習（Manifold learning）
*   降維
>高維度彎曲複雜的結構-->簡單形式呈現於低維度空間
*   處理非線性的複雜資料結構
*   比較 LLE VS PCA
>LLE適合<font color=#FF0000>非線性</font>	資料結構降維
>
>PCA適合<font color=#FF0000>線性</font>	資料結構降維



```python
from sklearn.datasets import samples_generator
from sklearn.manifold import LocallyLinearEmbedding


data, color = samples_generator.make_swiss_roll(n_samples=1500)
n_neighbors = 12 # 鄰近點數 
n_components = 2 # 降維後的維度
model = LocallyLinearEmbedding(n_neighbors=n_neighbors,
n_components=n_components)
model.fit(data)
x = model.transform(data)
print(x) # 轉換後的資料
```
