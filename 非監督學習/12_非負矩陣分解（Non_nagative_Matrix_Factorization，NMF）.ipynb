#非負矩陣分解（Non-nagative Matrix Factorization，NMF）



*   降維
>矩陣分解方式之一
>
>電腦視覺化（Computer Vision）、文字探勘（Text Mining）、推薦系統（Recommendation System）
*   原始矩陣所有元素皆為非負數
*   分解後矩陣的所有元素皆為非負數
*   潛在語義空間的各維度不一定正交

V = W * H
V1 ≈ W1 * H


```python
from sklearn.decomposition import NMF
from sklearn.datasets.samples_generator import make_blobs


centers = [[5, 10, 5], [10, 4, 10], [6, 8, 8]]
X, y = make_blobs(centers=centers) # centers為中心產生100筆數據
n_components = 2 # 潛在變數量
model = NMF(n_components=n_components)
model.fit(X)
W = model.transform(X) # 分解後
H = model.components_
print(W)
#print(H)

```
