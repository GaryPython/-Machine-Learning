# 隱含狄利克雷分佈（Latent Dirichlet Alloocation，LDA）



*   降維
>建立文件模型
>
>處理自然語言

步驟：


1.   隨機賦予個文件裡的詞彙一個主題
2.   依據分配給詞彙的主題，算出各文件的「主題機率」
3.   依據分配給詞彙的主題，算出各主題的「詞彙機率」
4.   根據2. 和 3.相乘後的機率，再度賦予「各文件的詞彙主題」
5.   重複 2. 3. 4. ，直到滿足收斂條件




補充：
  
[LDA主题－20newsgroups](https://blog.csdn.net/TiffanyRabbit/article/details/76445909)

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
 

# remove文本以外的資訊
data = fetch_20newsgroups(remove=('headers', 'footers', 'quotes'))
max_features = 1000
# 文字資料轉換為向量
tf_vectorizer = CountVectorizer(max_features=max_features,
stop_words='english')
tf = tf_vectorizer.fit_transform(data.data)
#20個主題
n_topics = 20
model = LatentDirichletAllocation(n_components=n_topics)
model.fit(tf)
print(model.components_) # 各主題詞彙分佈
#print(model.transform(tf)) #以主題呈現的文件
```

##補充 scikit-learn TFIDF


1.   詞頻（term frequency，TF）
2.   逆向檔案頻率（inverse document frequency，IDF）

目的：


*   找到文件當中的關鍵字，<font color=#FFFF00>怎樣的關鍵字是重要的？</font>	
  ><font color=#FFFF00>TF</font>	
  >
  >   出現最多次的字-><font color=#FF0000>不過因為每個文件的字數不同，無法比較</font>	
  >> 文件內的字數作為分母，將所有文件得到數值加以規範化。這也就是TF，特定單字在文件出現次數/文件總次數
  >
  ><font color=#FFFF00>IDF</font>
  >
  >定冠詞或是常用單字，得到很高的分數，<font color=#FF0000>a, an, the, and, or 等等</font>	
  >>1.先去除這些字詞(stopword)
  >>
  >>2.IDF會去計算一個字出現在文件的逆向頻率，這表示<font color=#FF0000>出現頻率越高</font>，出現在越多文件之中，但是<font color=#FF0000>得分會越低</font>

*  <font color=#FFFF00>TF和IDF相乘：TFxIDF，得到的綜合分數就是TFIDF。</font>



```python
#套件
from sklearn.feature_extraction.text import CountVectorizer
#「CountVectorizer」會計算單字出現在文件的次數；再透過「TfidfVectorizer」轉換成TFIDF和IDF
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
#直接使用「TfidfTransformer」計算TFID
import pandas as pd
```

###建立測試假文


```python
d1 = 'a b d e d f a f e fa d s a b n'
d2 = 'a z a f e fa h'
d3 = 'a z a f e fa h'
```

###計算：




*   <font color=#FF0000>vectorizer.get_feature_names()</font>	可以取得計算的單字
*   <font color=#FF0000>token_pattern</font>是<font color=#FF0000>(?u)\\b\\w\\w+\\b</font>，會過濾掉兩個字母以下的內容，但測試文本使用單個字母來測試，所以要加以改寫。將<font color=#FF0000>stop_word</font>設為<font color=#FF0000>None</font>也是同樣道理，比免去除單字，因為只是範例，而想看看所有結果：



####計算次數使用“<font color=#FF0000>CountVectorizer</font>”

補充：
[中英文操做](https://blog.csdn.net/weixin_38278334/article/details/82320307)

```python
vectorizer = CountVectorizer(stop_words=None, token_pattern="(?u)\\b\\w+\\b")  
X = vectorizer.fit_transform([d1,d2,d3])
r = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names(),index=['d1', 'd2', 'd3'])

print("CountVector")
r

```

####計算TFIDF使用“<font color=#FF0000>TfidfVectorizer</font>”

```python
vectorizer = TfidfVectorizer(sublinear_tf=False, stop_words=None, token_pattern="(?u)\\b\\w+\\b", smooth_idf=True)  
X = vectorizer.fit_transform([d1,d2,d3])
r = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names(), index=['d1', 'd2', 'd3'])
print("TFIDF")
r
```

```python
print("IDF")
pd.DataFrame([vectorizer.idf_], columns=vectorizer.get_feature_names())
```

####計算TFIDF使用“<font color=#FF0000>TfidfTransformer</font>“

```python
transformer = TfidfTransformer(smooth_idf=True)
Z = transformer.fit_transform(X)
r = pd.DataFrame(Z.toarray(),columns=vectorizer.get_feature_names(), index=['d1', 'd2', 'd3'])
print("TFIDF")
r
```

```python
print("IDF")
pd.DataFrame([transformer.idf_], columns=vectorizer.get_feature_names())
```
